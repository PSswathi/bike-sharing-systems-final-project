```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Example of loading your dataset, ensure the DataFrame `df` includes columns: 'temp', 'humidity', 'casual', and 'registered'
# df = pd.read_csv('path_to_your_bike_sharing_data.csv')

# Select relevant features and target variables
features = ['temp', 'humidity']  # Add other features as needed
target_casual = 'casual'
target_registered = 'registered'

# Split the data for predicting 'casual' rentals
X_casual = df[features]
y_casual = df[target_casual]
X_train_casual, X_test_casual, y_train_casual, y_test_casual = train_test_split(X_casual, y_casual, test_size=0.2, random_state=42)

# Split the data for predicting 'registered' rentals
X_registered = df[features]
y_registered = df[target_registered]
X_train_registered, X_test_registered, y_train_registered, y_test_registered = train_test_split(X_registered, y_registered, test_size=0.2, random_state=42)

# Standardize the features for neural network training
scaler = StandardScaler()
X_train_casual = scaler.fit_transform(X_train_casual)
X_test_casual = scaler.transform(X_test_casual)

X_train_registered = scaler.fit_transform(X_train_registered)
X_test_registered = scaler.transform(X_test_registered)

# Ensure the data is in correct format (NumPy arrays)
X_train_casual = np.array(X_train_casual)
X_test_casual = np.array(X_test_casual)
y_train_casual = np.array(y_train_casual)
y_test_casual = np.array(y_test_casual)

X_train_registered = np.array(X_train_registered)
X_test_registered = np.array(X_test_registered)
y_train_registered = np.array(y_train_registered)
y_test_registered = np.array(y_test_registered)

# Function to build a neural network model
def build_nn_model(input_shape):
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=(input_shape,)))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))  # Output layer
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# Build and train the model for 'casual' users
model_casual = build_nn_model(X_train_casual.shape[1])
model_casual.fit(X_train_casual, y_train_casual, epochs=50, batch_size=16, validation_split=0.2, verbose=1)

# Build and train the model for 'registered' users
model_registered = build_nn_model(X_train_registered.shape[1])
model_registered.fit(X_train_registered, y_train_registered, epochs=50, batch_size=16, validation_split=0.2, verbose=1)

# Predict and evaluate for 'casual' users
y_pred_casual = model_casual.predict(X_test_casual)
mse_casual = mean_squared_error(y_test_casual, y_pred_casual)
r2_casual = r2_score(y_test_casual, y_pred_casual)

# Predict and evaluate for 'registered' users
y_pred_registered = model_registered.predict(X_test_registered)
mse_registered = mean_squared_error(y_test_registered, y_pred_registered)
r2_registered = r2_score(y_test_registered, y_pred_registered)

# Display evaluation metrics
print(f"Casual Users - MSE: {mse_casual:.2f}, R: {r2_casual:.2f}")
print(f"Registered Users - MSE: {mse_registered:.2f}, R: {r2_registered:.2f}")
```
